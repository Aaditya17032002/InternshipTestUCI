```python
import pandas as pd # for data loading and cleaning.
import matplotlib.pyplot as plt #for map plotting
import seaborn as sns #for attractive, statistical plots with less code

df = pd.read_csv("taxi_zone_lookup.csv")

print(df.head())
print(df.info())
print(df.describe())

print(df.isnull().sum())

df = df.dropna()

df = df.drop_duplicates()

plt.figure(figsize=(10, 5))
sns.countplot(data=df, x='Borough', order=df['Borough'].value_counts().index)
plt.title("Number of Zones per Borough")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

plt.figure(figsize=(10, 6))
df['Zone'].value_counts().head(10).plot(kind='bar')
plt.title("Top 10 Most Common Zones")
plt.ylabel("Count")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

plt.figure(figsize=(6, 4))
sns.countplot(data=df, x='service_zone')
plt.title("Distribution of Service Zones")
plt.tight_layout()
plt.show()


plt.figure(figsize=(6, 4))
sns.heatmap(df[['LocationID']].corr(), annot=True)
plt.title("Correlation Heatmap")
plt.show()


df.to_csv("cleaned_taxi_zone_data.csv", index=False)
```
